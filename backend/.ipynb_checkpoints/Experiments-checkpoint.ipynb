```python
model = models.get_model("vgg16", cuda = False)
dataset = datasets.ImageNetSubset(root = "C:\\ImageNet")
loader = datasets.get_data_loader(dataset, batch_size = 1, shuffle = False)
explainer = XAI_core.ConceptExplainer(model, *dataset.mean_std())
```

```python
layer = model._modules["classifier"][3]
```

```python
x,y = dataset[30]
print(y)
x = x[None,:]
x_viz = dataset.prep_for_viz(x)
plt.axis("off")
plt.imshow(x_viz[0])
```

```python
from zennit.image import imgify, imsave

gamma = 1000
```

```python
composite_args = dict(epsilon = 1e-6, gamma = gamma)
explainer = XAI_core.ConceptExplainer(model, *dataset.mean_std(), composite_args = composite_args)

_, pred, grad = explainer.explain(x, target_logits = torch.tensor([30]).long(),  layer = layer)
print(pred)
grad = grad.sum(axis = 1).cpu()
imgify(grad, symmetric = True, cmap='coldnhot')
```

```python
composite_args = dict(epsilon = 1e-6, gamma = gamma)
explainer = XAI_core.ConceptExplainer(model, *dataset.mean_std(), composite_args = composite_args)

_, _, grad = explainer.explain(x, target_logits = torch.tensor([y]).long())
grad = grad.sum(axis = 1).cpu()
imgify(grad, symmetric = True, cmap='coldnhot')
```
